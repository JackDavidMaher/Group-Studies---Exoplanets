```python
!pip install pandas
import pandas as pd
import numpy as np
#read in csv file using panda
df = pd.read_csv('PS_2026.01.15_09.23.48.csv', comment='#')

#label each header with the parameters needed 
exo_table = df[['pl_name', 'pl_orbper', 'pl_orbsmax', 'pl_masse', 'pl_rade',
                'pl_trandep', 'pl_trandur', 'pl_orbeccen', 'pl_orbincl', 'st_rad', 'st_mass', 'st_teff', 'pl_ratror', 'sy_jmag']]

exo_table.head()

#filtering the mini neptunes from the planet radius, values can be easily adjusted
mini_nep = exo_table[
    (exo_table['pl_rade'] >= 1.3) &
    (exo_table['pl_rade'] <= 2.8)
]

#table of mini neptunes
mini_nep.head()
#print(mini_nep)
```

```python
mini_nep.to_csv('out.mini_neptune')
```

```python
#effective temperature 

#no value for albedo on exoplant archive, assume albedo = 0.3 (commonly used estimate for modelling rocky exoplanets)
A = 0.3
#units of stellar radius in solar radii, convert into km 
mini_nep['st_rad_m'] = mini_nep['st_rad'] * 695700

#units of semi major axis in AU, convert into km so theyre in the same units as the stellar radius
mini_nep['pl_orbsmax_km'] = mini_nep['pl_orbsmax']*1.496e+8
#use these values for eq temp

#equilibrium temperature formula from slides
mini_nep['pl_eqtemp'] = (
    (1 - A)**(1/4)
    * np.sqrt(mini_nep['st_rad_m'] / (2 * mini_nep['pl_orbsmax_km']))
    * mini_nep['st_teff']
)


mini_nep_temp = mini_nep[
    (mini_nep['pl_eqtemp'] >=200) &
    (mini_nep['pl_eqtemp'] <=1500)
]

#table of mini neptunes 
print(mini_nep_temp)
```

```python
mini_nep_temp.to_csv('out.mini_neptune_eqtemp')
```
